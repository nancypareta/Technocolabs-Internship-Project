```python
df.loc[df['ID'].isin(dupe_ids),:].head(10)
```

```python
#Similarly we can apply Masking for other condtions as well :
zero_mask_df=df==0
```

```python
#Calculating sum of 0 masked features , except ID column :
feat_zero_mask_df=zero_mask_df.iloc[:,1:].all(axis=1)
sum(feat_zero_mask_df)
```

```python
#Using tild symbol , we can check out Non-zero values in our dataframe :
df_clean_1=df.loc[~feat_zero_mask_df].copy()
```

```python
df_clean_1
```

```python
df['PAY_1'].unique()
```

```python
df_clean_1[df_clean_1['PAY_1']=='Not available'].count()
```

```python
na_mask_pay1=df['PAY_1']!='Not available'
```

```python
#Applying Not available boolean mask , to remove text feature from PAY 1
df_clean_2=df_clean_1.loc[na_mask_pay1,:].copy()
```

```python
df_clean_2.head()
```

```python
df_clean_2.info()
```

```python
#Converting Object or text to numeric 'int ' data type :
df_clean_2['PAY_1']=df_clean_2['PAY_1'].astype('int64')
```

```python
df_clean_2.info()
```

```python
df_clean_2['EDUCATION'].value_counts()
```

```python
#Replacing Non important values with important values :
df_clean_2['EDUCATION'].replace(to_replace=[0, 5, 6], value=4, inplace=True)
df_clean_2['EDUCATION'].value_counts()
```

```python
df_clean_2['MARRIAGE'].value_counts()
```

```python
#Similarly with marraige
df_clean_2['MARRIAGE'].replace(to_replace=0, value=3, inplace=True)
df_clean_2['MARRIAGE'].value_counts()
```

```python
#Making A new Column -> EDUCATION_CAT
df_clean_2['EDUCATION_CAT'] ='none'
```

```python
df_clean_2[['EDUCATION', 'EDUCATION_CAT']].head(10)
```

```python
#Defining A dictionary To Map Values for different categories :
cat_mapping = {
    1: "graduate school",
    2: "university",
    3: "high school",
    4: "others"
}


df_clean_2['EDUCATION_CAT'] = df_clean_2['EDUCATION'].map(cat_mapping)
df_clean_2[['EDUCATION', 'EDUCATION_CAT']].head(10)
```

```python
#Now let us get one hot encoded values for EDUCATION_CAT for training Our Model :
edu_ohe = pd.get_dummies(df_clean_2['EDUCATION_CAT'])
edu_ohe.head(10)
```

```python
df_with_ohe = pd.concat([df_clean_2, edu_ohe], axis=1)
df_with_ohe
```

```python
df_with_ohe[['EDUCATION_CAT', 'graduate school',
             'high school', 'university', 'others']].head(10)
```

```python
#Importing A cleaned Format of Our Excel File to CSV File for future references... :
df_with_ohe.to_csv('cleaned_data.csv', index=False)
```
